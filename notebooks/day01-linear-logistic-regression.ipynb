### 实战应用

结合两个算法进行综合分析：
1. 线性回归分析趋势方向
2. 逻辑回归分析涨跌概率
3. 综合决策制定交易策略

```python
# 综合分析函数
def comprehensive_analysis():
    """综合线性回归和逻辑回归的分析结果"""
    
    # 1. 线性回归分析趋势
    linear_result = main_linear_regression()
    
    # 2. 逻辑回归分析涨跌概率
    logistic_result = main_logistic_regression()
    
    # 3. 生成综合报告
    print("\n🎯 综合分析报告:")
    print("=" * 50)
    
    if linear_result["success"] and logistic_result["success"]:
        print("✅ 两个算法均执行成功")
        
        # 获取趋势方向
        trend_weight = linear_result["training_results"]["metrics"]["weight"]
        trend = "上涨" if trend_weight > 0 else "下跌"
        
        # 获取涨跌概率
        accuracy = logistic_result["training_results"]["test_metrics"]["accuracy"]
        
        print(f"📈 趋势方向: {trend} (权重: {trend_weight:.4f})")
        print(f"🎯 预测准确率: {accuracy:.2%}")
        print(f"📊 模型置信度: {'高' if accuracy > 0.7 else '中' if accuracy > 0.6 else '低'}")
        
        # 交易建议
        if trend_weight > 0 and accuracy > 0.65:
            print("💡 交易建议: 考虑买入")
        elif trend_weight < 0 and accuracy > 0.65:
            print("💡 交易建议: 考虑卖出或观望")
        else:
            print("💡 交易建议: 市场不明朗，建议观望")
    else:
        print("❌ 算法执行失败，请检查日志")
    
    print("=" * 50)

# 运行综合分析
comprehensive_analysis()
```
